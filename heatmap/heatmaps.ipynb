```python
import numpy as np
import imageio.v3 as iio
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

def create_custom_heatmap(input_path, output_path):
    # Load image
    image = iio.imread(input_path)
    
    # Convert image to grayscale if it's not already
    if len(image.shape) == 3:
        image = np.mean(image, axis=2)
    
    # Normalize image to range [0, 1]
    image = image / np.max(image)
    
    # Apply Gaussian filter to smooth the image
    smooth_image = gaussian_filter(image, sigma=4)
    
    # Create custom colormap (blue background with red-yellow gradient for high values)
    def custom_colormap(x):
        # Define the background color (blue)
        bg_color = np.array([0, 0, 1])
        
        # Define the gradient colors
        colors = []
        for val in x:
            if val < 0.5:  # Background
                colors.append(bg_color)
            elif val < 0.7:  # Transition to yellow
                t = (val - 0.5) / 0.2
                color = np.array([t, t, 1 - t])
                colors.append(color)
            else:  # Transition to red
                t = (val - 0.7) / 0.3
                color = np.array([1, 1 - t, 0])
                colors.append(color)
        return np.array(colors)
    
    # Create figure
    plt.figure(figsize=(6.4, 4.8))
    
    # Apply threshold to create mask
    threshold = 0.3  # Adjust this value to control sensitivity
    mask = smooth_image < threshold
    smooth_image[mask] = 0
    
    # Create custom colormap
    x = np.linspace(0, 1, 256)
    colors = custom_colormap(x)
    custom_cmap = plt.cm.colors.ListedColormap(colors)
    
    # Plot with custom colormap
    plt.imshow(smooth_image, cmap=custom_cmap, interpolation='bilinear')
    plt.axis('off')
    
    # Set background color to blue
    plt.gca().set_facecolor('blue')
    plt.gcf().set_facecolor('blue')
    
    # Save the heatmap
    plt.savefig(output_path, 
                bbox_inches='tight', 
                pad_inches=0.1, 
                dpi=600, 
                transparent=False,
                facecolor='blue')
    
    plt.show()

# Use the function
input_path = '/Users/sinan/Downloads/HSC.png'
output_path = '/Users/sinan/Desktop/CE_Exp_Hm.png'
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
from matplotlib.gridspec import GridSpec

def create_tracking_visualization(tracking_data, sigma=4, bins=50):
    """
    Create a combined visualization with 2D heatmap and marginal histograms
    
    Parameters:
    tracking_data: 2D numpy array of shape (n_points, 2) containing x,y coordinates
    sigma: smoothing factor for the heatmap
    bins: number of bins for histograms
    """
    # Create figure with custom layout
    fig = plt.figure(figsize=(12, 10))
    gs = GridSpec(3, 3, figure=fig)
    
    # Create the main heatmap
    ax_heatmap = fig.add_subplot(gs[1:, :-1])
    ax_histx = fig.add_subplot(gs[0, :-1])
    ax_histy = fig.add_subplot(gs[1:, -1])
    
    # Generate 2D histogram data
    heatmap, xedges, yedges = np.histogram2d(
        tracking_data[:, 0], 
        tracking_data[:, 1], 
        bins=bins
    )
    
    # Apply Gaussian smoothing to the heatmap
    heatmap_smooth = gaussian_filter(heatmap.T, sigma=sigma)
    
    # Plot the heatmap
    extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
    im = ax_heatmap.imshow(
        heatmap_smooth,
        extent=extent,
        origin='lower',
        cmap='gnuplot2',
        aspect='auto'
    )
    
    # Create color bar
    plt.colorbar(im, ax=ax_heatmap, label='Time spent (arbitrary units)')
    
    # Create marginal histograms
    ax_histx.hist(tracking_data[:, 0], bins=bins, color='blue', alpha=0.7)
    ax_histy.hist(tracking_data[:, 1], bins=bins, orientation='horizontal', color='blue', alpha=0.7)
    
    # Set labels
    ax_heatmap.set_xlabel('X Position')
    ax_heatmap.set_ylabel('Y Position')
    ax_histx.set_ylabel('Counts')
    ax_histy.set_xlabel('Counts')
    
    # Remove unnecessary ticks from histograms
    ax_histx.tick_params(labelbottom=False)
    ax_histy.tick_params(labelleft=False)
    
    # Align the histograms with the heatmap
    ax_histx.set_xlim(ax_heatmap.get_xlim())
    ax_histy.set_ylim(ax_heatmap.get_ylim())
    
    plt.suptitle('Tracking Analysis: Position and Time Spent', y=0.95)
    plt.tight_layout()
    
    return fig

# Example usage:
# Assuming your tracking data is stored in a numpy array with shape (n_points, 2)
# where each row contains [x, y] coordinates

def generate_sample_data(n_points=1000):
    """Generate sample circular trajectory data for demonstration"""
    t = np.linspace(0, 10*np.pi, n_points)
    noise = np.random.normal(0, 0.1, (n_points, 2))
    r = 2 + noise[:, 0]*0.2
    x = r * np.cos(t)
    y = r * np.sin(t)
    return np.column_stack([x + noise[:, 0], y + noise[:, 1]])

# Generate sample data
sample_data = generate_sample_data()

# Define paths
input_path = '/Users/sinan/Downloads/HSC.png'  # Update this path based on your actual path
output_path = '/Users/sinan/Downloads/H_SP_Hm.png'  # Path where the heatmap will be saved


# Create and save the visualization
fig = create_tracking_visualization(sample_data, sigma=2, bins=50)
plt.savefig('tracking_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import imageio.v3 as iio
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

# Define paths
input_path = '/Users/sinan/Downloads/2.png'  # Update this path based on your actual path
output_path = '/Users/sinan/Downloads/C_SP1_Hm.png'  # Path where the heatmap will be saved

# Load the image
image = iio.imread(input_path)

# Convert the image to grayscale if it's not already
if len(image.shape) == 3:  # If image has multiple channels (e.g., RGB)
    # Use weighted conversion to grayscale
    image = np.dot(image[..., :3], [0.2989, 0.5870, 0.1140])  # RGB to grayscale

# Normalize the image to range [0, 1]
image = (image - np.min(image)) / (np.max(image) - np.min(image) + 1e-7)  # Avoid division by zero

# Apply a Gaussian filter to smooth the image (optional, adjust sigma as needed)
smooth_image = gaussian_filter(image, sigma=30)

# Create the heatmap using the 'gnuplot2' colormap
plt.figure(figsize=(6.4, 4.8))  # Set the figure size
plt.imshow(smooth_image, cmap='gnuplot2', interpolation='bilinear')

# Remove axis ticks and labels for a cleaner output
plt.xticks([])
plt.yticks([])

# Save the heatmap to the output path
plt.savefig(output_path, bbox_inches='tight', pad_inches=0, dpi=600)
plt.show()
```
